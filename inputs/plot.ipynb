```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded CSV
file_path = '../outputs/predictions_20250807_214817.csv'
df = pd.read_csv(file_path)

# Extract genus_species from the Prediction column (first two parts)
df['Genus_Species'] = df['Prediction'].apply(lambda x: "_".join(x.split('_')[:2]))

# Count detections per genus_species
species_counts_genus = df['Genus_Species'].value_counts()

# Plot
plt.figure(figsize=(12, 10))
species_counts_genus.plot(kind='bar')
plt.xlabel('Species', fontsize=16)
plt.ylabel('Number of detections', fontsize=16)
plt.title('Number of detections per species', fontsize=18)
plt.xticks(rotation=75, fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```
